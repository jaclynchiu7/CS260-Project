```python
# imports
import statistics
import itertools
from datetime import datetime
from scipy.stats import zscore
from tabulate import tabulate 
import pandas as pd
import numpy as np

```

```python
# upload csv files 
from google.colab import files
uploaded = files.upload()
```

```python
# convert to data frame 
import io
model_attributes = pd.read_csv(io.BytesIO(uploaded['models.csv']))
full_ratings = pd.read_csv(io.BytesIO(uploaded['ratings.csv']))
```

```python
ratings = full_ratings.drop(['Timestamp', 'Age', 'Occupation (ex. student, accountant, engineer)', "How much time you spend outdoors?"], axis = 1)
ratings.set_index('Subject')
```

```python
# z-transformation within subject
ratings_T = ratings.transpose()
ratings_T.columns = ratings_T.iloc[0]
ratings_T = ratings_T.iloc[1: , :]
ratings_T.index.name = "Ad_Num"
ratings_zscore = ratings_T.apply(zscore)
```

```python
# calculating mean and sd per ads 
ratings_mean = ratings_T.mean(axis=1)
ratings_sd = ratings_T.std(axis = 1)
ratings_zscore_mean = ratings_zscore.mean(axis=1)
ratings_zscore_sd = ratings_zscore.std(axis=1)
```

```python
model_attributes = model_attributes.astype({"Model": str})
model_attributes['ratings_mean'] = ratings_mean.array
model_attributes['ratings_sd'] = ratings_sd.array
model_attributes['ratings_zscore_mean'] = ratings_zscore_mean.array
model_attributes['ratings_zscore_sd'] = ratings_zscore_sd.array
model_attributes
```

```python
#calculating average ratings by model number
all_models = model_attributes['Model'].unique()
by_model_summary = pd.DataFrame(index = all_models)

group_model1 = model_attributes.groupby(by = ['Model'])
by_model_summary['mean rating']= round(group_model1['ratings_mean'].mean(),2)
by_model_summary['zscore rating'] = round(group_model1['ratings_zscore_mean'].mean(),2)
by_model_summary['std rating'] = round(group_model1['ratings_sd'].std(),2)
by_model_summary = by_model_summary.sort_index(ascending = True)
by_model_summary['learning rate'] = ['2e-6', '2e-6', '3e-6', '2e-6', '1e-6', '1e-6', 'NA']
by_model_summary['max training steps'] = ['800', '1500', '800', '800', '800', '800', 'NA']
by_model_summary['# of training images'] = ['10', '10', '10', '7', '5', '10', 'NA']
by_model_summary.index.name = 'Model Number'
by_model_summary
```

```python
#calculating average ratings by instance prompt  
All_instance = model_attributes['Instance'].unique()
by_instance_summary = pd.DataFrame(index=All_instance)

group_model2 = model_attributes.groupby(by = ['Instance'])
by_instance_summary['mean rating'] = round(group_model2['ratings_mean'].mean(),2)
by_instance_summary['zscore rating'] = round(group_model2['ratings_zscore_mean'].mean(),2)
by_instance_summary['std rating'] = round(group_model2['ratings_sd'].std(),2)
by_instance_summary
#here other is the advertisement, we can take out if needed 
```
